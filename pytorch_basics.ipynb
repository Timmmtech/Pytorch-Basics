Linear Regression with Gradient Descent

```python
import numpy as np
import torch

```

```python
# input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43],
              [91, 88, 64],
              [87, 134, 58],
              [102, 43, 37],
              [69, 96, 70]], dtype='float32')

```

```python
# Targets (apples, oranges)
targets = np.array([[56, 70],
                    [81, 101],
                    [119, 133],
                    [22, 37],
                    [103, 119]], dtype='float32')
```

```python
# convert inputs and targets to tensors
inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)
print(inputs)
print(targets)
```

Linear Regression Model

```python
# Weights and biases
w = torch.randn(2, 3, requires_grad=True)
b = torch.randn(2, requires_grad=True)
print(w)
print(b)
```

```python
# defining the model
def model(x):
  return x @ w.t() + b
```

```python
# Generate prediction
preds = model(inputs)
print(preds)
```

```python
# Compare with targets
print(targets)
```

Loss Function

```python
# MSE loss
def mse(t1, t2):
  diff = t1 - t2
  return torch.sum(diff * diff) / diff.numel()
```

```python
# Compute loss
loss = mse (preds, targets)
print(loss)
```

```python
# Compute gradient
loss.backward()
```

```python
# Gradients for weights
print(w)
print(w.grad)
```

```python
print(b)
print(b.grad)
```

Adjust weights and biases to reduce the loss

```python
with torch.no_grad():
  w -= w.grad * 1e-5
  b -= b.grad * 1e-5
```

```python
# verify that the loss is actually lower
# Ensure inputs and targets are torch.Tensors
if isinstance(inputs, np.ndarray):
    inputs = torch.from_numpy(inputs)
if isinstance(targets, np.ndarray):
    targets = torch.from_numpy(targets)

preds = model(inputs)
loss = mse(preds, targets)
print(loss)
```

```python
# Adjust weights & reset gradients
with torch.no_grad():
  w -= w.grad * 1e-5
  b -= b.grad * 1e-5
  w.grad.zero_()
  b.grad.zero_()
```

```python
print(w)
print(b)
```

```python
# Calculate loss with the new weights and biases
preds = model(inputs)
loss = mse(preds, targets)
print(loss)
```

```python
# Train for 100 epochs
for i in range (100):
  preds = model(inputs)
  loss = mse(preds, targets)
  loss.backward()
  with torch.no_grad():
    w -= w.grad * 1e-5
    b -= b.grad * 1e-5
    w.grad.zero_()
    b.grad.zero_()
```

```python
# Calculate loss again
preds = model(inputs)
loss = mse(preds, targets)
print(loss)
```

Dataset and DataLoader

```python
from torch.utils.data import TensorDataset
```

```python
# Define dataset
train_ds = TensorDataset(inputs, targets)
train_ds[0:4]
```

Create DataLoader

```python
from torch.utils.data import DataLoader
```

```python
# Define data loader
batch_size = 5
train_d1 = DataLoader(train_ds, batch_size, shuffle=True)
```

```python
for xb, yb in train_d1:
  print(xb)
  print(yb)
  break
```

nn.Linear

```python
# Define model
model = nn.Linear(3, 2)
print(model.weight)
print(model.bias)
```

```python
# Parameters
list(model.parameters())
```

```python
# Generate predictions
preds = model(inputs)
preds
```

Loss Function

```python
# Import nn,functional
import torch.nn.functional as r
```

```python
# Define loss function
loss_fn = r.mse_loss
```

```python
loss = loss_fn(model(inputs), targets)
print(loss)
```

Optimizer

```python
# Define optimizer
opt = torch.optim.SGD(model.parameters(), lr=1e-5)
```

Training the data

```python
# utility function to train the model
def fit(num_epochs, model, loss_fn, opt, train_dl):
  # Repeat for given number of epochs
  for epoch in range(num_epochs):
    # Train with batches of data
    for xb, yb in train_dl:
      # 1. Generate predictions
      pred = model(xb)
      #2. Calculate loss
      loss = loss_fn(pred, yb)
      # 3. Compute gradients
      loss.backward()
      #4. Update parameters using gradients
      opt.step()
      #5. Reset the gradients to zero
      opt.zero_grad()
      # Print the progress
      if (epoch+1) % 10 == 0:
        print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, loss.item()))
```

```python
fit(100, model, loss_fn, opt, train_d1)
```
